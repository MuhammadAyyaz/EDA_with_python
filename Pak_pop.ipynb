```python
# DETERMINING THE CHANGE IN POPULATION FROM 1998 TO 2017
IN_2017 = df['ALL SEXES (URBAN)'].sum()+df['ALL SEXES (RURAL)'].sum()
IN_1998 = df['POPULATION 1998 (URBAN)'].sum()+df['POPULATION 1998 (RURAL)'].sum()
print(f'The total population of Pakistan in 1998 was {IN_1998}')
print(f'The total population of Pakistan in 2017 was {IN_2017}')
print(f'The popolation was increased by {IN_2017-IN_1998} persons in 20 years on an average of {(IN_2017-IN_1998)/ 20} annually')
print(f'The population grown by {round((((IN_2017-IN_1998)*100)/IN_1998)/12,3)} % annually')
```
