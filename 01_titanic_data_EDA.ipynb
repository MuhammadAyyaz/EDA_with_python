```python
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
```

```python
# Binning of age column in 7 categories
bins = [0,2,5,12,18,30,50,80]
labels = ['infants','toddlers','Kids','Teenagers','Youngs','Middle Aged','Old']
df['binned_age'] = pd.cut(df['age'], bins = bins, labels = labels)
```

```python
df['binned_age'].value_counts()
```

```python
sns.histplot(df['binned_age'])
```

```python
# renaming a column
df.rename(columns={'binned_age': 'age_groups'},inplace = True)
df.head()
```

```python
# determining survival of passengers according to their age groups
df.groupby('survived')['age_groups'].value_counts()
```
